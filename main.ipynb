## Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([2023,2024,2025,2026])
y1=np.array([15,35,30,20])
y2=np.array([19,25,47,10])
line_style=dict(marker=".",
                ms=15,
                mfc='#c96f84',
                mec='#c96f84',
                ls='solid',
                lw=2,
)
plt.plot(x,y1,c="#3449eb",**line_style)
plt.plot(x,y2,c='purple',**line_style)
plt.show()


#ms=markersize
#mfc=markerfacecolor
#mec=markeredgecolor
#ls=linestyle
#lw=linewidth
#c=color
```

### Labels

```python
x=np.array([2023,2024,2025,2026])
y1=np.array([15,35,30,20])
y2=np.array([19,25,47,10])
line_style=dict(marker=".",
                ms=15,
                mfc='#c96f84',
                mec='#c96f84',
                ls='solid',
                lw=2,)

plt.title("Class size", fontsize=20,
                        family='Times New Roman' ,
                        fontweight= "bold",
                        color='black')

plt.xlabel("Year", fontsize=20,
                   fontweight="bold",
                   family='Times New Roman')
plt.ylabel("Students", fontsize=20,
                   fontweight="bold",
                   family='Times New Roman')

plt.xticks(x)

plt.plot(x,y1,c="#3449eb",**line_style)
plt.plot(x,y2,c='purple',**line_style)
plt.show()
```

### Grid Lines

```python
x=[1,2,3,4,5]
y=[5,10,15,20,25]


plt.grid(axis="y",linewidth=2,color='grey',linestyle='dashed')
plt.plot(x,y)
plt.show()
```

### Barchart

```python
Categories=np.array(["Grains",'Vegetebles','Nuts','Sweets','Fruits'])
Values=np.array([4,2,5,6,9])

plt.bar(Categories,Values,color='#c96f84')
# plt.barh(Categories,Values,color='#c96f84')
plt.show()
```

### Pie Charts

```python
categories=np.array(["Freshmen","Sophomores","Juniors","Seniors"])
values=np.array([300,270,250,225])
colors=["#d46e9d","#c45287","#c4417e","#c92c75"]
plt.pie(values,labels=categories,autopct="%1.2f",colors=colors,explode=[0,0,0,0.1],startangle=90,shadow=True)
plt.show()
```

### Histograms

```python
scores=np.random.normal(loc=80,scale=10,size=100)
scores=np.clip(scores,0,100)
plt.hist(scores,bins=10,
         edgecolor="black",
         color="#c96f84")

plt.title("Exam Scores")
plt.xlabel("Scores")
plt.ylabel("Students")

plt.show()
```

### SubPlots

```python
x=np.array([1,2,3,4,5])
figure,axes=plt.subplots(2,2)

axes[0,0].plot(x,x*2,color="red")
axes[0,0].set_title("x*2")

axes[0,1].plot(x,x**2,color="green")
axes[0,1].set_title("x**2")

axes[1,0].plot(x,x**3,color="blue")
axes[1,0].set_title("x**3")

axes[1,1].plot(x,(x**4)-2,color="black")
axes[1,1].set_title("(x**4)-2")

plt.tight_layout()
plt.show()
```
